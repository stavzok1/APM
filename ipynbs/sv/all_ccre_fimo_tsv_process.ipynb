```python
import subprocess

# 1. Check if conda run can find vep
result = subprocess.run(
    ["conda", "run", "--no-banner", "-n", "venp_env", "which", "vep"],
    capture_output=True, text=True
)
print("which vep stdout:", result.stdout)
print("which vep stderr:", result.stderr)
print("returncode:", result.returncode)

# 2. Check if it's maybe mamba instead of conda
result2 = subprocess.run(
    ["which", "conda"],
    capture_output=True, text=True
)
print("\nwhich conda:", result2.stdout)

result3 = subprocess.run(
    ["which", "mamba"],
    capture_output=True, text=True
)
print("which mamba:", result3.stdout)

# 3. Check if the env actually exists
result4 = subprocess.run(
    ["conda", "env", "list"],
    capture_output=True, text=True
)
print("\nconda envs:\n", result4.stdout)
```

```python
import subprocess
from pathlib import Path

vcf_dir = Path("/home/stavz/masters/gdc/APM/data/SV/raw_somatic_sv_test")
out_dir = Path("/tmp/vep_test")
out_dir.mkdir(exist_ok=True)

# Grab the first VCF
vcf_file = next(vcf_dir.glob("*.vcf"))
out_vcf = out_dir / vcf_file.name

cmd = [
    "vep",
    "--input_file", str(vcf_file),
    "--output_file", str(out_vcf),
    "--offline", "--cache",
    "--fasta", "/home/stavz/.vep/Homo_sapiens.GRCh38.dna.primary_assembly.fa",
    "--species", "homo_sapiens",
    "--assembly", "GRCh38",
    "--format", "vcf", "--vcf",
    "--force_overwrite", "--everything",
    "--symbol", "--variant_class", "--check_ref",
    "--fork", "4",
]

print("Running:", " ".join(cmd))
result = subprocess.run(cmd, capture_output=True, text=True)
print("returncode:", result.returncode)
print("STDOUT:", result.stdout[-500:] if result.stdout else "(empty)")
print("STDERR:", result.stderr[-500:] if result.stderr else "(empty)")
```
